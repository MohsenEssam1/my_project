### 7 - clean ` age` column 
> tip : note the type of age column , and NaN values
 - try to use `to_numeric` 
 - tip : you should choose which best value to put in place of these NaNs ( mean, median, ....etc)

```python
df['Q3: AGE'].value_counts()
df['Q3: AGE'].isna().sum()
df['Q3: AGE'].unique()
df['Q3: AGE'].replace({ 'Old enough','I can remember when Java was a cool new language','ancient','OLD','Many','MY NAME JEFF','45-55','old enough','no','46 Halloweens.','?',312,'hahahahaha','See question 2','Enough','older than dirt','24-50','Over 50',1000},np.nan,inplace=True)
```

```python
mean=df['Q3: AGE'].mean()
std=df['Q3: AGE'].std()
nans=df['Q3: AGE'].isna().sum()
rand=np.random.randint(mean-2*std,mean+2*std,size=nans)
df['Q3: AGE'][np.isnan(df['Q3: AGE'])]=rand

```

### 8 - clean ` country` column
> tip : note that country column has country names with different writing ways 
  - use fuzzywuzzy to deal with this problem
  - you should choose which best value to put in place of these NaNs ( mode, unknown,other ....etc)

```python
df['Q4: COUNTRY'].replace({'subscribe to dm4uz3 on youtube','Europe','Earth','Fear and Loathing',"I don't know anymore"},np.nan,inplace=True)
df['Q4: COUNTRY'].unique()
```

```python
unique_val=df['Q4: COUNTRY'].unique()
category=["USA","Canada","America",'United States','UK','Us']
for state in category:
  country=process.extract(state,df['Q4: COUNTRY'],limit=len(unique_val))
  for c_country in country:
     if c_country[1]>=60:
         df.loc[df['Q4: COUNTRY']==c_country[0],'Q4: COUNTRY']=state
df['Q4: COUNTRY'].value_counts()
```

```python
df['Q4: COUNTRY'].replace({'United States','USA USA USA!!!!','I pretend to be from Canada, but I am really from the United States',"USAUSAUSA",'America','USA? Hard to tell anymore..','u s a','u.s.','U.S.','Us','U.S.A.'},"USA",inplace=True)
df['Q4: COUNTRY'].replace({'United Kingdom'},"UK",inplace=True)
df['Q4: COUNTRY'].value_counts()
df['Q4: COUNTRY'].describe()
df['Q4: COUNTRY'].fillna("USA",inplace=True)
df['Q4: COUNTRY'].value_counts()


```

### 9 - clean ` area` column
> tip : area column has the same problem as country columns , but looking at all it's unique values may give us another way to deal with it rather than try to fix it

```python
unique_val=df['Q5: STATE, PROVINCE, COUNTY, ETC'].unique()
category=['California','New York','NY']
for state in category:
  country=process.extract(state,df['Q5: STATE, PROVINCE, COUNTY, ETC'],limit=len(unique_val))
  for c_country in country:
     if c_country[1]>=60:
         df.loc[df['Q5: STATE, PROVINCE, COUNTY, ETC']==c_country[0],'Q5: STATE, PROVINCE, COUNTY, ETC']=state
df['Q5: STATE, PROVINCE, COUNTY, ETC'].value_counts()
```

```python
from pandas.io.formats.style_render import Subset
df['Q5: STATE, PROVINCE, COUNTY, ETC'].replace({'New York'},"NY",inplace=True)
df['Q5: STATE, PROVINCE, COUNTY, ETC'].unique()
df['Q5: STATE, PROVINCE, COUNTY, ETC'].replace({'USA','America','ur mom'},np.nan,inplace=True)
df.dropna(subset=['Q5: STATE, PROVINCE, COUNTY, ETC'],inplace=True)
```
